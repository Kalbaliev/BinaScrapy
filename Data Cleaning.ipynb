```python
def replaceTime(val):
    currentDay = datetime.now().day
    currentMonth = datetime.now().month
    currentYear=datetime.now().year
    
    yesterday= (datetime.now() - timedelta(days=1)).day
    yesterdayMonth= (datetime.now() - timedelta(days=1)).month
    yesterdayYear= (datetime.now() - timedelta(days=1)).year
    months={"Yanvar":"01","Fevral":"02","Mart":"03","Aprel":"04","May":"05","İyun":"06","İyul":"07","Avqust":"08","Sentyabr":"09","Oktyabr":10,"Noyabr":11,"Dekabr":12}
    
    if len(str(currentMonth))==1:
        currentMonth=f'0{currentMonth}'
    if len(str(yesterdayMonth))==1:
        yesterdayMonth=f'0{yesterdayMonth}'
    val=val.replace("Yeniləndi:","").strip()
    if "Dünən" in val:
        val=f'{yesterday} {yesterdayMonth} {yesterdayYear}'
    elif "Bugün" in val:
        val=f'{currentDay} {currentMonth} {currentYear}'
    for monthname, monthnum in months.items():
        #print(monthname)
        if monthname in val:
            #print("-----",monthname)
            val=val.replace(monthname, monthnum)
    return val.replace(" ","")

data["UPDATETIME"]=data.UPDATETIME.apply(lambda x: replaceTime(x))
```
